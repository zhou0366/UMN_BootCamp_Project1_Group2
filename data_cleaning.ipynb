```python
#dependencies
import pandas as pd
```

```python
#load data file
water_qual_df = pd.read_csv("waterPollution.csv")
water_qual_df.head()
```

```python
#make dataframes for measurements of mg{O2}/L, pH, or Cel
water_qual_summary_df = water_qual_df[(water_qual_df["resultUom"] == "mg{O2}/L")|(water_qual_df["resultUom"] == "[pH]")|(water_qual_df["resultUom"] == "Cel")]

water_qual_summary_df.head()
```

```python
#narrow down columns to resultUom, phenomenonTimeReferenceYear, resultMeanValue, Country, PopulationDensity, netMigration_2011_2018, gdp, waterBodyIdentifier

#empty dataframe
clean_df = pd.DataFrame()
#add columns
clean_df["Country"] = water_qual_summary_df["Country"]
clean_df["Water Body ID"] = water_qual_summary_df["waterBodyIdentifier"]
clean_df["Year"] = water_qual_summary_df["phenomenonTimeReferenceYear"]
clean_df["Unit of Measure"] = water_qual_summary_df["resultUom"]
clean_df["Value"] = water_qual_summary_df["resultMeanValue"]
clean_df["Population Density"] = water_qual_summary_df["PopulationDensity"]
#clean_df["Net Migration 2011-2018"] = water_qual_summary_df["netMigration_2011_2018"]
clean_df["GDP"] = water_qual_summary_df["gdp"]


clean_df.head()
```

```python
#narrow values of phenomenonTimeReferenceYear column 2011-2018 as this is where the netmigration is measured
#cutting net migration stat
#clean_df = clean_df[(clean_df["Year"] >= 2011) & (clean_df["Year"] <= 2018)]
#clean_df.head()
```

```python
#additional cleanup for empty rows, etc
clean_df.dropna(subset=['Country'], inplace = True)
```

```python
#export cleaned csv
clean_df.reset_index()
clean_df.to_csv("cleaned_data.csv")
```
